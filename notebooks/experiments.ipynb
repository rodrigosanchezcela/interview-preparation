```python
# Check for special characters in chunks
import re

def analyze_text_quality(chunks, name):
    print(f"\n{name} TEXT QUALITY ANALYSIS:")
    print("="*80)
    
    bullet_count = 0
    special_char_count = 0
    avg_length = 0
    
    for chunk in chunks:
        text = chunk.text
        avg_length += len(text)
        
        # Check for bullet points
        if any(marker in text for marker in ['•', '◦', '▪', '–', '―']):
            bullet_count += 1
        
        # Check for excessive special characters
        special_chars = len(re.findall(r'[^\w\s.,!?;:()\'-]', text))
        if special_chars > 5:
            special_char_count += 1
    
    avg_length = avg_length / len(chunks) if chunks else 0
    
    print(f"Total chunks: {len(chunks)}")
    print(f"Average chunk length: {avg_length:.1f} characters")
    print(f"Chunks with bullet points: {bullet_count} ({bullet_count/len(chunks)*100:.1f}%)")
    print(f"Chunks with excessive special chars: {special_char_count} ({special_char_count/len(chunks)*100:.1f}%)")
    
analyze_text_quality(selected_cv_chunks, "CV")
analyze_text_quality(selected_job_chunks, "JOB")
```

## Optimization Plan

**Problems Identified:**
1. 83% of CV chunks contain bullet points (•) - noise for embeddings
2. CV chunks too short (149 chars) vs Job chunks (415 chars) - granularity mismatch
3. No text cleaning before embedding - bullets, section names repeated
4. Job chunks incomplete (cut off mid-sentence: "Awareness of")
5. Chunking strategy too aggressive - splits related content

**Solutions:**
1. **Text Cleaning**: Remove bullets, normalize whitespace, clean special chars
2. **Better Chunking**: Combine related bullet points into semantic groups
3. **Normalize Granularity**: Make CV and Job chunks similar size
4. **Sentence Completion**: Don't cut mid-sentence

## Addressing the Core Problem: Granularity Mismatch

**Current Situation:**
- CV chunks: Individual bullet points (~149 chars) - e.g., "• Deployed chatbot using Docker"
- Job chunks: Full requirement descriptions (~400 chars) - e.g., "Technical Skills: Python, AWS, Docker..."
- **Problem**: Comparing a single bullet point to a full job requirement loses context

**Solution: Group Related CV Bullet Points**
- Combine all bullets from the same Experience/Education entry into ONE chunk
- Example: Instead of 5 separate bullets about one job, create 1 chunk with all 5 bullets
- This creates CV chunks with similar context richness as job chunks

**Expected Result:**
- CV chunks become ~300-500 chars (similar to job chunks)
- More context = better semantic matching
- A full project description can now match against full job requirements

```python
# Step 1: Group CV chunks by section and consecutive items
# The idea: bullets under the same heading belong together

def group_cv_chunks_by_context(chunks, max_length=500):
    """
    Group consecutive bullet points from the same section into single chunks.
    
    Why? A job posting under "Experience" with 4 bullet points describes ONE role.
    We want to treat those 4 bullets as a single semantic unit.
    
    Args:
        chunks: List of CVChunk objects
        max_length: Maximum characters per grouped chunk
    
    Returns:
        List of grouped chunks with richer context
    """
    grouped = []
    current_group = []
    current_section = None
    current_length = 0
    
    for chunk in chunks:
        # Only group ITEM type (bullet points)
        if chunk.chunk_type != "ITEM":
            continue
        
        # Start new group if section changes or max length exceeded
        if (current_section != chunk.section or 
            current_length + len(chunk.text) > max_length):
            
            # Save previous group
            if current_group:
                combined_text = " ".join([c.text for c in current_group])
                grouped.append({
                    'text': combined_text,
                    'section': current_section,
                    'original_chunks': len(current_group)
                })
            
            # Start new group
            current_group = [chunk]
            current_section = chunk.section
            current_length = len(chunk.text)
        else:
            # Add to current group
            current_group.append(chunk)
            current_length += len(chunk.text)
    
    # Don't forget last group
    if current_group:
        combined_text = " ".join([c.text for c in current_group])
        grouped.append({
            'text': combined_text,
            'section': current_section,
            'original_chunks': len(current_group)
        })
    
    return grouped

# Apply grouping
grouped_cv_chunks = group_cv_chunks_by_context(selected_cv_chunks, max_length=500)

print(f"Original CV chunks: {len(selected_cv_chunks)}")
print(f"Grouped CV chunks: {len(grouped_cv_chunks)}")
print(f"\nCompression ratio: {len(selected_cv_chunks)/len(grouped_cv_chunks):.1f}x")
print("(e.g., 5 bullets → 1 group means 5x compression)")
```

```python
# Step 2: Inspect grouped chunks to see the difference
print("\nGROUPED CV CHUNKS EXAMPLES:")
print("="*80)

for i in range(min(3, len(grouped_cv_chunks))):
    group = grouped_cv_chunks[i]
    print(f"\nGroup {i+1}:")
    print(f"  Section: {group['section']}")
    print(f"  Combined {group['original_chunks']} bullet points")
    print(f"  Length: {len(group['text'])} chars")
    print(f"  Text preview: {group['text'][:250]}...")
    print("-"*80)

# Compare chunk lengths
import numpy as np
original_lengths = [len(c.text) for c in selected_cv_chunks]
grouped_lengths = [len(g['text']) for g in grouped_cv_chunks]

print(f"\nCHUNK SIZE COMPARISON:")
print(f"  Original avg: {np.mean(original_lengths):.0f} chars (individual bullets)")
print(f"  Grouped avg:  {np.mean(grouped_lengths):.0f} chars (combined bullets)")
print(f"  Job chunks avg: ~{np.mean([len(c.text) for c in job_chunks_selected_sentences]):.0f} chars")
print("\n→ Grouped CV chunks now match job chunk size! ✓")
```

```python
# Step 3: Generate embeddings for grouped chunks
print("\nGenerating embeddings for grouped CV chunks...")

# Convert grouped dicts to text list for embedding
grouped_texts = [group['text'] for group in grouped_cv_chunks]

# Need to create temporary chunks for the embedder
# (it expects CVChunk objects with .text attribute)
from interview_prep.schemas.cv_schema import CVChunk

temp_chunks = []
for i, group in enumerate(grouped_cv_chunks):
    temp_chunks.append(CVChunk(
        section_id=i,
        text=group['text'],
        section=group['section'],
        chunk_type="GROUPED_ITEM",
        chunk_id=i,
        location=0
    ))

cv_embeddings_grouped = embedder_clean.calculate_embeddings_chunks(temp_chunks)
print(f"✓ Grouped CV embeddings: {cv_embeddings_grouped.shape}")
```

```python
# Step 4: Compare all three approaches
similarity_grouped = matcher.match(cv_embeddings_grouped, job_embeddings_sentences)

print("="*80)
print("FINAL COMPARISON: Three Approaches")
print("="*80)

# Approach 1: Original (individual bullets, word-split jobs)
orig_mean = similarity_matrix.mean()
orig_max = similarity_matrix.max()

# Approach 2: Text cleaning + sentence-split jobs
clean_mean = similarity_sentences.mean()
clean_max = similarity_sentences.max()

# Approach 3: Grouped CV + sentence-split jobs
group_mean = similarity_grouped.mean()
group_max = similarity_grouped.max()
group_best = similarity_grouped.max(axis=0)

print(f"\n1. BASELINE (individual bullets, word-split):")
print(f"   Mean: {orig_mean:.3f} | Max: {orig_max:.3f}")

print(f"\n2. CLEANING + SENTENCE-SPLIT:")
print(f"   Mean: {clean_mean:.3f} | Max: {clean_max:.3f}")
print(f"   Improvement: {((clean_mean - orig_mean) / orig_mean * 100):+.1f}%")

print(f"\n3. GROUPED CV + SENTENCE-SPLIT:")
print(f"   Mean: {group_mean:.3f} | Max: {group_max:.3f}")
print(f"   Improvement: {((group_mean - orig_mean) / orig_mean * 100):+.1f}%")
print(f"   Best per job: {group_best}")
print(f"   Jobs with >0.5: {(group_best > 0.5).sum()}/{len(group_best)}")

print(f"\n{'='*80}")
print(f"WINNER: Approach #{[orig_mean, clean_mean, group_mean].index(max([orig_mean, clean_mean, group_mean])) + 1}")
print(f"{'='*80}")
```
